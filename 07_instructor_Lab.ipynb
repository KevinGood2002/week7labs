```python
gh_subsample = gh_teams.sample(frac=0.5, random_state=42)  # 50% random subsample with reproducibility
gh_subsample
```

```python

kf = KFold(n_splits=8, shuffle=True, random_state=42)  # 8-fold cross-validation with shuffling

for fold, (train_index, test_index) in enumerate(kf.split(gh_teams)):
    print(f"Fold {fold+1}")
    train_set = gh_teams.iloc[train_index]
    test_set = gh_teams.iloc[test_index]
    print(f"Train size: {len(train_set)}, Test size: {len(test_set)}\n")
```
